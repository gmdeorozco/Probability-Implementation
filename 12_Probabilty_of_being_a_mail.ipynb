# 1. Problem: In Orange County, 51% of the adults are males. One adult is randomly selected for a survey involving credit card usage. It is later learned that the selected survey subject was smoking a cigar. Also, 9.5% of males smoke cigars, whereas 1.7% of females smoke cigars (based on data from the Substance Abuse and Mental Health Services Administration). Use this additional information to find the probability that the selected subject is a male.

# 3. Programing Solution

```python
def prob_person_being_a_male(
    prob_being_a_male: float=0.51,
    prob_smoking_given_you_are_a_male:float = 0.095,
    prob_smoking_given_you_are_a_woman:float=0.017 ):

  prob_male = (prob_being_a_male * prob_smoking_given_you_are_a_male) / (
    prob_being_a_male * prob_smoking_given_you_are_a_male +
    (1 - prob_being_a_male) * prob_smoking_given_you_are_a_woman
)
  return prob_male


# Data:
prob_being_a_male = 0.51
prob_smoking_given_you_are_a_male = 0.095
prob_smoking_given_you_are_a_woman = 0.017

print(prob_person_being_a_male())

```
