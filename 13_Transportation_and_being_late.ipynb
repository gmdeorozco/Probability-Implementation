# 1. Problem: A person uses his car 30% of the time, walks 30% of the time and rides the bus 40% of the time as he goes to work. He is late 10% of the time when he walks; he is late 3% of the time when he drives; and he is late 7% of the time he takes the bus. a. -> What is the probability he took the bus if he was late?  b.-> What is the probability he walked if he is on time?

# 3. Programing Solution

### 3.1 Data

```python
#Data

uses_car = 0.30
uses_bus = 0.40
walks = 0.30

is_late_when_walks = 0.10
is_late_when_car = 0.03
is_late_when_bus = 0.07

# Questions
# What is the probability he took the bus if he was late?
# What is the probability he walked if he is on time?
```

### 3.2 What is the probability he took the bus if he was late?

```python
def prob_took_the_bus_given_he_is_late(
    uses_car = 0.30,
    uses_bus = 0.40,
    walks = 0.30,

    is_late_when_walks = 0.10,
    is_late_when_car = 0.03,
    is_late_when_bus = 0.07,):
  return uses_bus * is_late_when_bus/(uses_car*is_late_when_car + walks * is_late_when_walks + uses_bus * is_late_when_bus)

print(f'Result: The probability he took the bus if he was late is -> {prob_took_the_bus_given_he_is_late()}')
```

### 3.3 What is the probability he walked if he is on time?

```python
def prob_walked_given_he_is_on_time(
    uses_car = 0.30,
    uses_bus = 0.40,
    walks = 0.30,

    is_late_when_walks = 0.10,
    is_late_when_car = 0.03,
    is_late_when_bus = 0.07,):

  return walks * (1-is_late_when_walks) / (uses_car*(1-is_late_when_car) + walks * (1-is_late_when_walks) + uses_bus * (1-is_late_when_bus))

print(f'Result: The probability he walked if he is on time is -> {prob_walked_given_he_is_on_time()}')
```
