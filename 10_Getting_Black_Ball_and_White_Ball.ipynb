# 1. Problem: If you have a bag with 10 black balls and 5 white balls, what is the probability of drawing a black ball on the first draw and a white ball on the second draw without replacement?

# 3. Programing Solution

```python
def prob_getting_black_ball_then_white_ball():
  # P(Black Ball) = #black_balls / total_balls
  prob_black_ball = 10/15

  #P(White Ball | Given Black ball was taken) = #white_balls / remaining_balls
  prob_white_ball = 5 /(15-1)

  prob_event = prob_black_ball * prob_white_ball
  return prob_event

print(prob_getting_black_ball_then_white_ball())
```
