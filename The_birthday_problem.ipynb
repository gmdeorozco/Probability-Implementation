# 1. Problem: In a classroom of 30 students, what is the probability that at least two students have the same birthday?

# 3. Programing Solution

```python
# P(Same day) = 1 - P(Different day)
# Calculating P(Different day):

def probability_of_having_different_birthday(n_students:int=30,n_days:int = 365) -> float:
  if n_students == 1:
    return 1
  n_days = n_days - 1
  return n_days/365 * probability_of_having_different_birthday(n_students - 1 , n_days )

# Now calculating P(Same day)
def probability_of_having_same_birthday(n_students:int = 30, n_days:int = 365):
  return round(1 - probability_of_having_different_birthday(n_students, n_days),2)

```

# 4. Testing

```python
prob = probability_of_having_same_birthday(n_students=30, n_days=365)
assert prob == 0.71 , f'Incorrect, it should be 0.71, returned: {prob}'

print('Probability computed correctly')
```
