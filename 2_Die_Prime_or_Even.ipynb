## 3. Helper Functions

### 3.1 Prime numbers
Return True if the number is Prime and False if it is not.

```python
# Returns True if a given number 'n' is Prime
def is_prime(n: int) -> bool:
  if n <= 1:
    return False

  for i in range(2,int(n**0.5)+1):
    if n % i == 0:
      return False
  return True

assert is_prime(2) == True, f'Incorrect, 2 is Prime, returned : {is_prime(2)}'
print("Prime condition returned correctly")
```

### 3.2 Count Prime numbers up to X


```python
def count_prime_numbers_up_to_x(x: int) -> int:
  count = 0
  for n in range(2,x+1):
    if is_prime(n):
      count += 1
  return count

assert count_prime_numbers_up_to_x(20) == 8, f'Incorrect, prime numbers up to 20 are 8, returned: {count_prime_numbers_up_to_x(20)}'
print("Counted prime numbers correctly")
```

### 3.3 Even Numbers up to X
The purpose of not countig '2' is because 2 even AND prime.

```python
def count_even_numbers_up_to_x(x: int) -> int:
  return x//2

assert count_even_numbers_up_to_x(21) == 10, f'Incorrect, even numbers up to 21 are 10, returned {count_even_numbers_up_to_x(21)}'
print("Counted even numbers, correctly")
```

### 3.4 Count Even and Prime Numbers up to X

```python
def count_prime_and_even_numbers_up_to_x(x: int ) -> int:
    count = 0
    for i in range(2,x+1):
        if is_prime(i) and i % 2 == 0:
            count += 1
    return count

assert count_prime_and_even_numbers_up_to_x(20) == 1, f'Incorrect, only 2 is even and prime up to 20, returned: {count_prime_and_even_numbers_up_to_x(20)}'
print('Counted numbers that are prime and even correctly')
```

# 3. Generalization Function

```python
#Generalization
def get_probability_of_even_or_prime_numbers_from_die( n: int=6 ) -> int:
    # P(E or Pr)=P(E)+P(Pr)P(E and Pr)

    prob_even = count_even_numbers_up_to_x(n)/n
    prob_prime = count_prime_numbers_up_to_x(n)/n
    prov_even_and_prime = count_prime_and_even_numbers_up_to_x(n)/n

    prob_even_or_prime = prob_even + prob_prime - prov_even_and_prime

    return prob_even_or_prime

```

# 4. Testing

```python
prob = get_probability_of_even_or_prime_numbers_from_die( n=6 )
assert  prob == 5/6, f'Incorrect it should be {5/6} returned: {get_probability_of_even_or_prime_numbers_from_die(n=6)}'
print('Probabiliy of even or prime numbers computed correctly')
```
