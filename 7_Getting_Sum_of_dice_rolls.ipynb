# 3. Programing Solution

### 3.1 Using a Formula to find the pairs that sum 7

```python
def get_number_of_pairs_that_sum_target(target:int, a=1, b=6):
  return max(0,min(target-a,b)-max(1,target-b)) + 1

print(get_number_of_pairs_that_sum_target(7))
```

### 3.2 Using bruteforce to find the pairs

```python
def find_by_brute_force(target:int=7):
  count = 0
  for x in range(1,7):
    for y in range(1,7):
      if x + y == target:
        count += 1
  return count

print(find_by_brute_force(7))
```

### 3.3 Testing Formula vs Bruteforce

```python
assert get_number_of_pairs_that_sum_target(7) == find_by_brute_force(7), f'Results are different, please review functions {get_number_of_pairs_that_sum_target(7)} vs { find_by_brute_force(7)}'
assert get_number_of_pairs_that_sum_target(8) == find_by_brute_force(8), f'Results are different, please review functions {get_number_of_pairs_that_sum_target(8)} vs { find_by_brute_force(8)}'
print('Both algorithms give the same result')
```

### 3.4 Now that we have the number of pairs, we calculate:

```python
def probability_to_obtain_sum_in_six_sided_dice(target:int = 7):
  all_probabilities = 6 ** 2
  sum_target_probability = get_number_of_pairs_that_sum_target(target = 7)
  return sum_target_probability / all_probabilities
```

# 4. Testing

```python
prob = probability_to_obtain_sum_in_six_sided_dice(7)
assert prob == 6/36, f'Incorrect, The probabilities for obtainig 7 is given by 6/36, returned: {prob}'

print('Probability computed correctly')
```
