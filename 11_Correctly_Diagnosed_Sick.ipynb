# 1. Problem: In a population of 1 000 000 individuals, a certain condition affects 1 of each 10 000 people. If a diagnostic test is 99% accurate, what is the probability that a randomly chosen person is correctly diagnosed as having the condition?

# 3. Programing Solution

```python
# P(Being sick|Diagnosed sick) = [ P(Diagnosed sick | Being sick)*P(Being sick) ] / P(Diagnosed sick)
def prob_correctly_dignosed_sick(prob_affected: float=1/10000, prob_accuracy: float=0.99, population:int=1000000) -> float:
  people_sick = population * prob_affected
  people_healthy = population - people_sick

  people_sick_diagnosed_sick = prob_accuracy * people_sick
  # people_sick_diagnosed_healthy = people_sick - prob_accuracy * people_sick

  # people_healthy_diagnosed_healthy = prob_accuracy * people_healthy
  people_healthy_diagnosed_sick = people_healthy - (prob_accuracy * people_healthy)

  prob_being_sick_given_diagnosed_sick = people_sick_diagnosed_sick / ( people_healthy_diagnosed_sick + people_sick_diagnosed_sick)

  return prob_being_sick_given_diagnosed_sick

print(prob_correctly_dignosed_sick(1/10000,0.99,1000000))
```
