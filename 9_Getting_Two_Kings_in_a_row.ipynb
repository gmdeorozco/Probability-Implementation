# 3. Programing Solution

```python
#P(Getting first king)
probability_of_getting_first_king = 4/52

#P(Getting second king | given that first king was gotten already)
probability_of_getting_second_king = ( 4-1 ) / 51

probability_of_getting_two_kings = probability_of_getting_first_king * probability_of_getting_second_king

print(f'Result : {probability_of_getting_two_kings}')
assert probability_of_getting_two_kings == (4/52)*(3/51),f'Incorrect, please check formula'
print('Probability computed correctly')
```
