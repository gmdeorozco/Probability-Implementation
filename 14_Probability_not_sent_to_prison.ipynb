# 1. Problem: In a study of pleas and prison sentences, it is found that 45% of the subjects studied were sent to prison. Among those sent to prison, 40% chose to plead guilty. Among those not sent to prison, 55% chose to plead guilty.

## a. If one of the study subjects is randomly selected, find the probability of getting someone who was not sent to prison.

## b. If a study subject is randomly selected and it is then found that the subject entered a guilty plead, find the probability that this person was not sent to prison.

# 3. Programing Solution

### 3.1 Response A: Probability not sent to prison

```python

def probability_not_sent_to_prison(
    sent_to_prison = 0.45,
plead_guilty_given_sent_to_prison = 0.40,
plead_guilty_given_not_sent_to_prison = 0.55,
):
  return 1 - sent_to_prison

print(f'Result: Probabiliy not sent to prison: {probability_not_sent_to_prison()}')
```

```python
def probability_not_sent_to_prison_given_entered_guilty_plead(
    sent_to_prison = 0.45,
    plead_guilty_given_sent_to_prison = 0.40,
    plead_guilty_given_not_sent_to_prison = 0.55,):
  return (1 - sent_to_prison) * plead_guilty_given_not_sent_to_prison/( (1 - sent_to_prison) * plead_guilty_given_not_sent_to_prison
                                                                       + sent_to_prison * plead_guilty_given_sent_to_prison)
print(f'Result: Probability not sent to prison given entered guilty plead {probability_not_sent_to_prison_given_entered_guilty_plead()}')
```
