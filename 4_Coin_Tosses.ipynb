# 1. Problem: Consider tossing a fair coin three times. What is the probability of getting heads on all three tosses? Remember, each toss is independent.

# 3. Programing Solution

```python
# P(HHH) = P(H)**n_tosses
def same_face_probability(n_tosses: int=3) -> float:
  prob_face = 1/2
  prob_all_same_face = prob_face**n_tosses
  return prob_all_same_face
```

4. Testing

```python
assert same_face_probability(n_tosses=3) == 1/8, f'Incorrect, the probability of getting same face 3 times is {1/8}, returned: {same_face_probability(3)}'
assert same_face_probability(n_tosses=4) == 1/16, f'Incorrect, the probability of getting same face 3 times is {1/16}, returned: {same_face_probability(4)}'
print('Computed same face probability correctly')
```
